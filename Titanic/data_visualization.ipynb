```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
from numpy.random import randn
```

```python
path_train = "E:\\python_things\\Titanic_csv\\Titanic\\train.csv"
data_train = pd.read_csv(path_train)
p = pd.DataFrame(data_train)
p.info()
p
```

```python
#存活的人数

p.Survived.value_counts().plot(kind="bar")
plt.xlabel("survived")
plt.ylabel("number")
plt.title("survivor(0 is death)")
```

```python
#船票价格的去曲线图
#df = pd.DataFrame.plot(p,columes=["Fare"],index=p.Survived)
#df.plot
y = p.Survived
x = p.Fare
p.sort_values(["Survived"],inplace=True)
plt.ylim(-1,2)
plt.yticks([-1,0,1,2])
plt.plot(x,y)
plt.xlabel("Fare")
plt.ylabel("survived of number")
plt.title("船票和存活的关系",fontproperties='SimHei',fontsize=20)
```

```python
#船票的和获救关系的散点图
plt.scatter(x=p.Survived,y=p.Fare)
plt.xlabel("Fare")
plt.ylabel("survived of numbers")
plt.title("船票和存活的关系",fontproperties='SimHei',fontsize=20)
```

```python
p.Pclass.value_counts().plot(kind="bar")
plt.xlabel("Pclass")
plt.ylabel("survived of number")
plt.title("船舱等级人数的统计",fontproperties='SimHei',fontsize=20)
```

```python
#船舱等级与存活关系
survived_0 = p.Pclass[p.Survived==0].value_counts()
survived_1 = p.Pclass[p.Survived==1].value_counts()
pd.DataFrame({'living':survived_0,'death':survived_1}).plot(kind="bar",stacked=True)
plt.xlabel("Pclass")
plt.ylabel("survived of number")
plt.title("船舱等级和存活的关系",fontproperties='SimHei',fontsize=20)
```
